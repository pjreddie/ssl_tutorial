## Training From Scratch

- Very small Vision Transformer (ViT)
- Train 30 epochs on CIFAR10

```python
class Tokenizer(nn.Module):
    def __init__(self, patch_size, dim, num_patches):
        super().__init__()
        self.patch_embed = nn.Conv2d(3, dim, kernel_size=patch_size, stride=patch_size)
        self.cls_token = nn.Parameter(torch.zeros(1, 1, dim))
        self.pos_embed = nn.Parameter(torch.zeros(1, num_patches + 1, dim))  # learned

    def forward(self, x):                   # (B, 3, 32, 32)
        x = self.patch_embed(x)             # (B, 128, 8, 8)
        x = x.flatten(2).transpose(1, 2)    # (B, 64, 128)
        # Add in class token + positional embeddings
        cls_tokens = self.cls_token.expand(len(x), -1, -1)
        x = torch.cat([cls_tokens, x], dim=1) + self.pos_embed  # (B, 65, 128)
        return x
```

```python
class Block(nn.Module):
    """One transformer block: attention + feed-forward, both with residuals."""
    def __init__(self, dim, heads, mlp_dim):
        super().__init__()
        self.norm1 = nn.LayerNorm(dim)
        self.attn = nn.MultiheadAttention(dim, heads, batch_first=True)
        self.norm2 = nn.LayerNorm(dim)
        self.mlp = nn.Sequential(
            nn.Linear(dim, mlp_dim), nn.GELU(), nn.Linear(mlp_dim, dim))

    def forward(self, x):
        h = self.norm1(x)
        x = x + self.attn(h, h, h, need_weights=False)[0]
        x = x + self.mlp(self.norm2(x))
        return x

class Encoder(nn.Module):
    def __init__(self, dim, depth, heads, mlp_dim):
        super().__init__()
        self.blocks = nn.Sequential(*[Block(dim, heads, mlp_dim) for _ in range(depth)])

    def forward(self, x):
        return self.blocks(x)
```

```python
class ViT(nn.Module):
    def __init__(self, image_size, patch_size, dim, depth, heads, mlp_dim, num_classes):
        super().__init__()
        num_patches = (image_size // patch_size) ** 2  # 64
        self.tokenizer = Tokenizer(patch_size, dim, num_patches)
        self.encoder = Encoder(dim, depth, heads, mlp_dim)
        self.classifier = nn.Linear(dim, num_classes)

    def forward(self, x):                               # (B, 3, 32, 32)
        x = self.tokenizer(x)                           # (B, 65, 128)
        x = self.encoder(x)
        return self.classifier(x[:, 0])                 # classify from the CLS token
```

```python
def accuracy(model, dataloader):
    model.to(device)
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in dataloader:
            outputs = model(images.to(device))
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels.to(device)).sum().item()
    return correct/total
```

```python
from utils import CosineWithWarmup

def train_classification(model, train_loader, val_loader, epochs, lr=0.002):
    model.to(device)
    model.train()
    total_steps = epochs*len(train_loader)
    warmup = 2000
    step=0
    loss_ema = None

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=0.05)
    scheduler = CosineWithWarmup(optimizer, warmup, total_steps)

    print("Training")
    for e in range(epochs):
        for data, labels in train_loader:
            optimizer.zero_grad()
            with torch.autocast(device_type=device.type, dtype=torch.bfloat16):
                output = model(data.to(device))
                loss = criterion(output, labels.to(device))
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            scheduler.step()
            step += 1

            if loss_ema is None:
                loss_ema = loss.item()
            else:
                loss_ema = .1*loss.item() + .9*loss_ema

            if(step % 100 == 0):
                print(f'{step}: {loss_ema}')
        print(f'Epoch {e} done, val accuracy: {accuracy(model, val_loader)}')
        model.train()
```

```python
from PIL import Image

def classify_image(model, image_path, target_size, classes):
    """
    Loads an image, preprocesses it, and classifies it using the given model.

    Args:
        model (torch.nn.Module): The trained classification model.
        image_path (str): Path to the image file.
        target_size (tuple): The (height, width) to resize the image to.
        classes (list): A list of class names corresponding to the model's output.

    Returns:
        str: The predicted class name.
    """
    image_transform = transforms.Compose([
        transforms.Resize(target_size),
        transforms.ToTensor(),
        transforms.Normalize(NORM_MEAN, NORM_STD)
    ])

    image = Image.open(image_path).convert('RGB')
    image_tensor = image_transform(image).unsqueeze(0)

    model.to(device)
    model.eval()

    with torch.no_grad():
        output = model(image_tensor.to(device))
        _, predicted = torch.max(output, 1)

    predicted_class_idx = predicted.item()
    return classes[predicted_class_idx]
```

# AutoEncoder

- Transformer encodes image
- Bottleneck restricts information flow
- Predict original pixels
- Reconstruction loss

![](https://raw.githubusercontent.com/pjreddie/ssl_tutorial/refs/heads/main/images/ae.png)

```python
class Detokenizer(nn.Module):
    """Inverse of Tokenizer: drop the class token, map each patch token back to pixels."""
    def __init__(self, patch_size, dim):
        super().__init__()
        self.patch_unembed = nn.ConvTranspose2d(dim, 3, kernel_size=patch_size, stride=patch_size)

    def forward(self, x):                   # (B, 65, dim)
        x = x[:, 1:]                        # throw out the class token
        side = int(x.shape[1] ** 0.5)       # 8 patches per side
        x = x.transpose(1, 2).reshape(len(x), -1, side, side)  # (B, dim, 8, 8)
        return self.patch_unembed(x)        # (B, 3, 32, 32)
```

```python
class ViTAutoEncoder(nn.Module):
    def __init__(self, image_size, patch_size, dim, depth, heads, mlp_dim, z_dim):
        super().__init__()
        num_patches = (image_size // patch_size) ** 2  # 64
        self.tokenizer = Tokenizer(patch_size, dim, num_patches)
        self.encoder = Encoder(dim, depth, heads, mlp_dim)

        # squeeze every token down to z_dim and back up
        self.bottleneck = nn.Sequential(
                            nn.Linear(dim, z_dim),
                            nn.Linear(z_dim, dim))

        self.decoder = Encoder(dim, 2, heads, mlp_dim)
        self.detokenizer = Detokenizer(patch_size, dim)

    def forward(self, x):                               # (B, 3, 32, 32)
        x = self.tokenizer(x)                           # (B, 65, 128)
        x = self.encoder(x)
        x = self.bottleneck(x)
        x = x + self.tokenizer.pos_embed                # tell the decoder where each token is
        x = self.decoder(x)
        return self.detokenizer(x)                      # (B, 3, 32, 32)
```

```python
def train_autoencoder(model, train_loader, epochs=1, lr=0.002, warmup=2000):
    model.to(device)
    model.train()
    total_steps = epochs*len(train_loader)
    step=0
    loss_ema = None

    criterion = nn.SmoothL1Loss(beta=.01)

    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=0.05)
    scheduler = CosineWithWarmup(optimizer, warmup, total_steps)

    print("Training")
    for e in range(epochs):
        for data, _ in train_loader:
            data = data.to(device)
            optimizer.zero_grad()
            with torch.autocast(device_type=device.type, dtype=torch.bfloat16):
                output = model(data)
                loss = criterion(output, data)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            scheduler.step()
            step += 1

            if loss_ema is None:
                loss_ema = loss.item()
            else:
                loss_ema = .1*loss.item() + .9*loss_ema

            if(step % 100 == 0):
                print(f'{step}: {loss_ema}')
        print(f'Epoch {e} done')
```

```python
from PIL import Image

def autoencode_and_display(model, image_path, target_size=(32, 32)):
    """
    Loads an image, preprocesses it, autoencodes it, and displays the original and reconstructed image.

    Args:
        model (torch.nn.Module): The trained autoencoder model.
        image_path (str): Path to the image file.
        target_size (tuple): The (height, width) to resize the image to.
    """
    # Preprocessing transformation
    image_transform = transforms.Compose([
        transforms.Resize(target_size),
        transforms.ToTensor(),
        transforms.Normalize(NORM_MEAN, NORM_STD)
    ])

    # Load and preprocess the original image
    original_image_pil = Image.open(image_path).convert('RGB')
    image_tensor = image_transform(original_image_pil).unsqueeze(0) # Add batch dimension

    # Move model to device and set to evaluation mode
    model.to(device)
    model.eval()

    # Perform autoencoding
    with torch.no_grad():
        reconstructed_tensor = model(image_tensor.to(device))

    # Denormalize and convert to PIL image for display
    # Undo normalization: img = img * std + mean
    reconstructed_tensor = reconstructed_tensor.squeeze(0).cpu()
    reconstructed_tensor = reconstructed_tensor * NORM_STD + NORM_MEAN
    reconstructed_pil = transforms.ToPILImage()(reconstructed_tensor.clamp(0, 1))

    # Convert original PIL image to tensor for imshow
    original_image_display_tensor = transforms.ToTensor()(original_image_pil)

    # Convert reconstructed PIL image to tensor for imshow
    reconstructed_display_tensor = transforms.ToTensor()(reconstructed_pil)

    # Display original and reconstructed images side-by-side
    print("Original Image:")
    imshow(original_image_display_tensor)
    print("Reconstructed Image:")
    imshow(reconstructed_display_tensor)

# Use the function with the dog image and the trained autoencoder
autoencode_and_display(vit_ae, image_path)
```

# Masked AutoEncoder

- Mask (hide) some parts of image
- Try to reconstruct hidde parts from visible
- Reconstruction loss on non-visible parts

![](https://raw.githubusercontent.com/pjreddie/ssl_tutorial/refs/heads/main/images/mae.png)

- https://arxiv.org/abs/2111.06377

```python
class ViTMaskedAutoEncoder(nn.Module):
    def __init__(self, image_size, patch_size, dim, depth, heads, mlp_dim):
        super().__init__()
        num_patches = (image_size // patch_size) ** 2  # 64
        self.tokenizer = Tokenizer(patch_size, dim, num_patches)
        self.encoder = Encoder(dim, depth, heads, mlp_dim)
        self.decoder = Encoder(dim, 2, heads, mlp_dim)
        self.detokenizer = Detokenizer(patch_size, dim)

    def forward(self, image, mask):                     # (B, 3, 32, 32)
        x = self.tokenizer(image)                       # (B, 65, 128)
        encoded = self.encoder(x[:, mask])

        # masked tokens come back as zeros, pos_embed will be added in
        full = torch.zeros_like(x)
        full[:, mask] = encoded

        x = full + self.tokenizer.pos_embed             # tell the decoder where each token is
        x = self.decoder(x)
        return self.detokenizer(x)                      # (B, 3, 32, 32)
```

```python
def train_masked_autoencoder(model, train_loader, epochs=1, lr=0.002, warmup=2000, mask_ratio=0.5):
    model.to(device)
    model.train()
    total_steps = epochs*len(train_loader)
    n_tokens = model.tokenizer.pos_embed.shape[1]  # 65
    f_side = int((n_tokens-1)**0.5)
    patch_size = model.tokenizer.patch_embed.kernel_size[0]  # pixels per patch side, 4
    step=0
    loss_ema = None

    criterion = nn.SmoothL1Loss(beta=.01, reduction='none')  # per-pixel, masked below
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=0.05)
    scheduler = CosineWithWarmup(optimizer, warmup, total_steps)

    print("Training")
    for e in range(epochs):
        for data, _ in train_loader:
            data = data.to(device)

            # 1: visible, 0: not visible
            mask = torch.rand(n_tokens, device=device) > mask_ratio

            optimizer.zero_grad()
            with torch.autocast(device_type=device.type, dtype=torch.bfloat16):
                output = model(data, mask)
                loss = criterion(output, data)          # per-pixel loss (B, 3, 32, 32)

                # ignore visible pixels in the loss
                visible = mask[1:].view(f_side, f_side)
                visible = visible.repeat_interleave(patch_size, 0).repeat_interleave(patch_size, 1)
                loss[:, :, visible] = 0
                loss = loss.mean()

            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            scheduler.step()
            step += 1

            if loss_ema is None:
                loss_ema = loss.item()
            else:
                loss_ema = .1*loss.item() + .9*loss_ema

            if(step % 100 == 0):
                print(f'{step}: {loss_ema}')
        print(f'Epoch {e} done')
```

# Contrastive Learning

- Create two different views of each image
- Run batch through the encoder
- Encodings of the same image should be the same, different from other images in a batch
- Contrastive loss
    - e.g. Classification loss: positive is the matching images, negatives are other images

![](https://raw.githubusercontent.com/pjreddie/ssl_tutorial/refs/heads/main/images/simclr.png)

- https://arxiv.org/abs/2002.05709
- https://arxiv.org/abs/2103.00020

```python
class ViTMaskedEncoder(nn.Module):
    def __init__(self, image_size, patch_size, dim, depth, heads, mlp_dim):
        super().__init__()
        num_patches = (image_size // patch_size) ** 2  # 64
        self.tokenizer = Tokenizer(patch_size, dim, num_patches)
        self.encoder = Encoder(dim, depth, heads, mlp_dim)

    def forward(self, image, mask):                     # (B, 3, 32, 32)
        x = self.tokenizer(image)                       # (B, 65, 128)
        return self.encoder(x[:, mask])
```

```python
def train_contrastive(model, train_loader, epochs=1, lr=0.002, warmup=2000, mask_ratio=0.5, temperature=0.1):
    model.to(device)
    model.train()
    total_steps = epochs*len(train_loader)
    n_tokens = model.tokenizer.pos_embed.shape[1]  # 65
    step=0
    loss_ema = None

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=0.05)
    scheduler = CosineWithWarmup(optimizer, warmup, total_steps)

    print("Training")
    for e in range(epochs):
        for data, _ in train_loader:
            data = data.to(device)

            # 1: visible, 0: not visible
            mask_a = torch.rand(n_tokens, device=device) > mask_ratio
            mask_b = torch.rand(n_tokens, device=device) > mask_ratio
            mask_a[0] = mask_b[0] = 1

            optimizer.zero_grad()
            with torch.autocast(device_type=device.type, dtype=torch.bfloat16):
                output_a = model(data, mask_a)[:, 0] # just the cls token
                output_b = model(data, mask_b)[:, 0]

                logits = F.normalize(output_a, dim=1) @ F.normalize(output_b, dim=1).T / temperature
                labels = torch.arange(len(data), device=device)
                loss = criterion(logits, labels)

            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            scheduler.step()
            step += 1

            if loss_ema is None:
                loss_ema = loss.item()
            else:
                loss_ema = .1*loss.item() + .9*loss_ema

            if(step % 100 == 0):
                print(f'{step}: {loss_ema}')
        print(f'Epoch {e} done')
```

# Latent Reconstruction

- Encode masked image, decode the hidden positions
- Separate "target encoder" encodes whole image
- Try to predict these targets from the masked image
- Reconstruction or contrastive loss (contrastive usually better)

![](https://raw.githubusercontent.com/pjreddie/ssl_tutorial/refs/heads/main/images/latentmim.png)

- https://arxiv.org/abs/2301.08243
- https://arxiv.org/abs/2407.15837
- https://arxiv.org/abs/2508.10104

```python
class ViTLatentMIM(nn.Module):
    def __init__(self, image_size, patch_size, dim, depth, heads, mlp_dim):
        super().__init__()
        num_patches = (image_size // patch_size) ** 2  # 64
        self.tokenizer = Tokenizer(patch_size, dim, num_patches)
        self.encoder = Encoder(dim, depth, heads, mlp_dim)
        self.decoder = Encoder(dim, 2, heads, mlp_dim)
        self.target_encoder = copy.deepcopy(self.tokenizer.patch_embed)

    def forward(self, image, mask):                     # (B, 3, 32, 32)
        x = self.tokenizer(image)                       # (B, 65, 128)
        encoded = self.encoder(x[:, mask])

        # masked tokens come back as zeros, pos_embed will be added in
        full = torch.zeros_like(x)
        full[:, mask] = encoded

        x = full + self.tokenizer.pos_embed             # tell the decoder where each token is
        return self.decoder(x)

    def get_targets(self, x):
        with torch.no_grad():
            x = self.target_encoder(x)      # (B, 128, 8, 8) EMA patch embeddings
            x = x.flatten(2).transpose(1, 2)  # (B, 64, 128)
            return x

    def update_target_encoder(self, a=.995):
        with torch.no_grad():
            self.target_encoder.weight.mul_(a).add_(self.tokenizer.patch_embed.weight, alpha=1-a)
            self.target_encoder.bias.mul_(a).add_(self.tokenizer.patch_embed.bias, alpha=1-a)
```

```python

def train_latent_mim(model, train_loader, epochs=1, lr=0.002, warmup=2000, mask_ratio=0.5, temperature=0.1):
    model.to(device)
    model.train()
    total_steps = epochs*len(train_loader)
    n_tokens = model.tokenizer.pos_embed.shape[1]  # 65
    step=0
    loss_ema = None

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=0.05)
    scheduler = CosineWithWarmup(optimizer, warmup, total_steps)

    print("Training")
    for e in range(epochs):
        for data, _ in train_loader:
            data = data.to(device)

            # 1: visible, 0: not visible
            mask = torch.rand(n_tokens, device=device) > mask_ratio
            mask[0] = 1                                 # need the cls token

            optimizer.zero_grad()
            with torch.autocast(device_type=device.type, dtype=torch.bfloat16):
                output = model(data, mask)              # (B, 65, 128) decoded tokens
                targets = model.get_targets(data)       # (B, 64, 128) EMA patch embeddings

                # contrastive loss 1 v all in image, ignore visible
                hidden = ~mask[1:]                      # the patch slots the model had to guess
                logits = F.normalize(output[:, 1:][:, hidden], dim=2) @ F.normalize(targets, dim=2).mT / temperature  # (B, n_hidden, 64)
                labels = hidden.nonzero().squeeze(1).expand(len(data), -1)  # each masked token's own patch index
                loss = criterion(logits.transpose(1, 2), labels)

            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            scheduler.step()
            model.update_target_encoder()

            step += 1

            if loss_ema is None:
                loss_ema = loss.item()
            else:
                loss_ema = .1*loss.item() + .9*loss_ema

            if(step % 100 == 0):
                print(f'{step}: {loss_ema}')
        print(f'Epoch {e} done')
```

```python
vit_latentmim = ViTLatentMIM(image_size=32, patch_size=4, dim=128, depth=5, heads=8, mlp_dim=512)
train_latent_mim(vit_latentmim, imagenet32['train'], 5, mask_ratio=0.75)
```
